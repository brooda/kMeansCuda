```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Get results

```python
size = 5000
```

```python
labels = open(f'../results/{size}.txt', 'r').read().split(",")
```

```python
labels = [int(l) for l in labels]
```

```python
colors= ['red' if l == 0 else 'yellow' if l == 1 else 'purple' if l == 2 else 'green' if l==3 else 'blue' for l in labels]
```

```python
df = pd.read_csv("./5/5000.csv", names=["x", "y", "z"])
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df.x, df.y, df.z, color=colors)
ax.set_xlim3d(-10, 10)
ax.set_ylim3d(-10, 10)
ax.set_zlim3d(-10, 10)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
